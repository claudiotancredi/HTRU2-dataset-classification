```python
import numpy as np
import GMM
import GMMDiag
import GMMTiedCov
import metrics
import PCA
import utils
```

```python
priors = [0.5, 0.9, 0.1]
D, L = utils.load('Train.txt')
ZD, mean, standardDeviation = utils.ZNormalization(D)
PCA7 = PCA.PCA(ZD, L, 7)
numberOfSplitToPerform = 6
```

```python
gmm = GMM.GMM()
minDCF3fold = []
print("Start GMM with 3-fold on z normalized features")
for i in range(len(priors)):
    print("")
    print("Working on application with prior:", priors[i])
    temp = utils.fastKfoldGMM(ZD, L, gmm, numberOfSplitToPerform, prior=priors[i])
    for c in range(numberOfSplitToPerform):
        print("For", 2**(c+1), "components and prior=", priors[i], "the minDCF of the GMM model is", temp[c])
        minDCF3fold.append(temp[c])
print("")
print("END")
```

```python
utils.plotDCFGMM([2**(c+1) for c in range(numberOfSplitToPerform)], minDCF3fold, "GMM components")
```

```python
gmm = GMMDiag.GMMDiag()
minDCF3foldDiag = []
print("Start GMM Diag with 3-fold on z normalized features")
for i in range(len(priors)):
    print("")
    print("Working on application with prior:", priors[i])
    temp = utils.fastKfoldGMMDiag(ZD, L, gmm, numberOfSplitToPerform, prior=priors[i])
    for c in range(numberOfSplitToPerform):
        print("For", 2**(c+1), "components and prior=", priors[i], "the minDCF of the GMM Diag model is", temp[c])
        minDCF3foldDiag.append(temp[c])
print("")
print("END")
```
