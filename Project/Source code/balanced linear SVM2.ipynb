```python
import utils
import metrics
import PCA
import balancedSVM
import matplotlib.pyplot as plt
import numpy as np
```

```python
priors = [0.5, 0.9, 0.1]
D, L = utils.load('Train.txt')
ZD, mean, standardDeviation = utils.ZNormalization(D)
C=np.logspace(-5, -2, num=15)
model = balancedSVM.SVM()
PCA7 = PCA.PCA(ZD, L, 7)
PCA6 = PCA.PCA(ZD, L, 6)
print ("Executing linear SVM with class re-balancing")
```

```python
# SINGLE FOLD Z NORMALIZED SVM LINEAR
(DTRSF, LTRSF), (DEVSF, LEVSF) = utils.split_db_singleFold(ZD, L)
minDCF = []
print("Start linear SVM on single fold of z normalized features")
for i in range(len(priors)):
    print("")
    print("Working on application with prior:", priors[i])
    for value in C:
        model.train(DTRSF, LTRSF, option='linear',C=value, piT=0.5) 
        temp = metrics.minimum_detection_costs(model.predictAndGetScores(DEVSF), LEVSF, priors[i], 1, 1)
        minDCF.append(temp)
        print("For C=", value, "the minDCF is", temp)
print("")
print("END")
```

```python
utils.plotDCF(C, minDCF, "C")
```
