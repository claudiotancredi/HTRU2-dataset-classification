```python
import utils
import metrics
import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np
import PCA
```

```python
priors = [0.5, 0.9, 0.1]
D, L = utils.load('Train.txt')
ZD, mean, standardDeviation = utils.ZNormalization(D)
lambdas=np.logspace(-5, 2, num=30)
lr = LogisticRegression.LogisticRegression()
PCA7 = PCA.PCA(ZD, L, 7)
PCA6 = PCA.PCA(ZD, L, 6)
```

```python
# SINGLE FOLD Z NORMALIZED LR
(DTRSF, LTRSF), (DEVSF, LEVSF) = utils.split_db_singleFold(ZD, L)
minDCF = []
print("Start logistic regression on single fold of z normalized features")
for i in range(len(priors)):
    print("")
    print("Working on application with prior:", priors[i])
    for l in lambdas:
        lr.train(DTRSF, LTRSF, l, 0.5) # pi_T = 0.5
        temp = metrics.minimum_detection_costs(lr.predictAndGetScores(DEVSF), LEVSF, priors[i], 1, 1)
        minDCF.append(temp)
        print("For lambda", l, "the minDCF is", temp)
print("")
print("END")
```

```python
utils.plotDCF(lambdas, minDCF, "λ")
```
